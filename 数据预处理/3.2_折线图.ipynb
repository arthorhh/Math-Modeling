```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%config InlineBackend.figure_format='svg'
# 字体设置
plt.rcParams['font.sans-serif'] = ['SimHei']    # SimHei黑体，Kaiti楷体，LiSu隶书，FangSong仿宋，YouYuan幼圆，STSong华文宋体
plt.rcParams['axes.unicode_minus'] = False
```

一、折线图

```python
x = np.arange(-50,51)
y = x ** 2
# 图表名称
plt.title("y等于x的平方")
#x轴y轴,fontsize设置文字大小，linewidth设置线条粗细,在plot里
plt.xlabel("x轴",fontsize=12)
plt.ylabel("y轴")
# 绘制线性图
plt.plot(x,y,linewidth=5)
plt.show()
```

```python
x = np.linspace(-2 * np.pi, 2 * np.pi, 120)
y = np.sin(x)
# 创建画布
plt.figure(figsize=(8, 4), dpi=120)
# 绘制折线图
plt.plot(x, y, linewidth=2, marker='*', color='red')
# 显示绘图
plt.show()
```

```python
x = np.arange(-10,10)
y1 = x ** 2
y2 = x
plt.title("y=x^2")    #设置背景板
plt.xlabel("x轴",fontsize=12)
plt.ylabel("y轴")
plt.plot(x,y1)    # 画线1
plt.plot(x,y2)    # 画线2
plt.show()
```

```python
times = ['2015/6/26','2015/8/1','2015/9/6','2015/10/12','2015/11/17','2015/12/23','2016/1/28','2016/3/4',
            '2016/5/15','2016/6/20','2016/7/26','2016/8/31','2016/10/6','2016/11/11','2016/12/17']
sales = np.random.randint(500,2000,size=len(times))    # 随机销量
# 直接plt.plot(times,sales)会导致x轴太挤
plt.xticks(range(1,len(times),2))    # 部分展示,xticks(tick=None列表,labels=None,**kwargs文本属性如旋转角度roration=45，颜色color=“red”)
plt.xticks(range(1,len(times),2),rotation=45)    # 斜体
plt.plot(times,sales)
plt.show()
```

```python
x = np.arange(-50,51)
y = x ** 2
plt.title("y=x^2 x取值范围为：[-10,10)",fontsize=16)
plt.xlabel('x轴',fontsize=12)
plt.ylabel('y轴')
x_titcks = [-50,-40,-30,-20,-10,0,10,20,30,40,50]
x_lables = ['%s度' %i for i in x_titcks]
plt.xticks(x_titcks,x_lables)
plt.plot(x,y)
plt.show()
```

```python
times = ['2015/6/26', '2015/8/1', '2015/9/6', '2015/10/12', '2015/11/17', '2015/12/23', '2016/1/28', '2016/3/4',
         '2016/5/15', '2016/6/20', '2016/7/26', '2016/8/31', '2016/10/6', '2016/11/11', '2016/12/17']
income = np.random.randint(500,2000,size=len(times))    # 随机收入
expense = np.random.randint(300,1500,size=len(times))    # 随机支出
plt.xticks(range(1,len(times),2),rotation=45)
plt.plot(times,income,label='收入1')    # 给线起一个名称叫收入1
plt.plot(times,expense,label='支出1')    # 给线起一个名称叫支出1
plt.legend()    # 把起的名称找个合适的地方放进去，即默认用每个图的label值作为图例的说明
#legend位置说明：[upper left,upper center,upper right,
#               center left,center,center right,
#               lower left,lower center,lower right]   loc='',默认best

plt.show()
```
